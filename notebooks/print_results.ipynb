## Expected Accuracy

```python
import os
import json
import pandas as pd

def get_acc(dataset_name: str, model_name: str) -> float:
    filepath = f"../outputs/{dataset_name}__{model_name}/ground_truth_summary.json"
    if not os.path.exists(filepath):
        return None
    with open(filepath, "r") as f:
        data = json.load(f)
    return data["mean_accuracy"]

dataset_names = ["triviaqa-validation", "simpleqa-verified", "gsm8k-test", "math-500", "aime25-test", "mmlu-validation", "gpqa-diamond"]
model_names = ["Olmo-3-7B-Instruct", "Qwen3-1_7B-non-thinking", "Qwen3-4B-non-thinking", "Qwen3-8B-non-thinking", "Qwen3-14B-non-thinking", "gpt-oss-20b"]

import pandas as pd

# Construct a DataFrame with model_names as rows, dataset_names as columns, and values are accs
acc_matrix = []
for model in model_names:
    row = []
    for dataset in dataset_names:
        acc = get_acc(dataset, model)
        row.append(acc)
    acc_matrix.append(row)

df = pd.DataFrame(acc_matrix, index=model_names, columns=dataset_names)
df
```

## P(True) or Verbalized Confidence

```python
import json

metric = "spearman_r"
method = "verbalized_confidence"  # "verbalized_confidence" or "ptrue"
isotonic = True
MODEL2GPU = {
    "Olmo-3-7B-Instruct": "RTXPro6000",
    "Qwen3-8B-non-thinking": "RTXPro6000",
    "gpt-oss-20b": "A40",
}
def get_metric(
    model_name: str,
    dataset_name: str,
    metric: str,
    isotonic: bool = False,
) -> float:
    filepath = f"../estimator_results/{method}/{dataset_name}__{model_name}/evaluation_metrics_cc.json"
    if isotonic:
        filepath = filepath.replace("cc.json", "cc_isotonic.json")
    with open(filepath, "r") as f:
        data = json.load(f)
    return data[metric]

model_names = ["Olmo-3-7B-Instruct", "Qwen3-8B-non-thinking", "gpt-oss-20b"]  #
dataset_names = ["triviaqa-validation", "simpleqa-verified", "gsm8k-test", "math-500", "aime-test", "mmlu-test", "gpqa-diamond"]
# dataset_names = ["triviaqa-train", "simpleqa-train", "gsm8k-train", "math-train", "aime-train", "mmlu-validation", "gpqa-train"]

import pandas as pd

matrix = []
for model in model_names:
    row = []
    for dataset in dataset_names:
        metric_value = get_metric(model, dataset, metric, isotonic)
        row.append(metric_value)
    matrix.append(row)

df = pd.DataFrame(matrix, index=model_names, columns=dataset_names)
df
```
